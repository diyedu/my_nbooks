```python
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
avocado = pd.read_csv('avocado.csv')
```

```python
avocado.head()
```

```python
avocado.info()
```

```python
avocado['Date'] = pd.to_datetime(avocado['Date'])
```

```python
avocado.info()
```

```python
albany_avo = avocado.query('year == 2015 and type == "conventional" and region == "Albany"')[['Date', 'Total Bags', 'Small Bags']]
```

```python
albany_avo.head()
```

```python
avocado_melted = avocado.melt(
    id_vars='Date',
    value_vars=['Total Bags', 'Small Bags'],
    var_name='Bags', value_name='Count')
```

```python
avocado_melted
```

```python
sns.relplot(data=avocado_melted, kind='line', x='Date', y='Count', hue='Bags')
```

```python
avocado['Sales Volume'] = pd.qcut(avocado['Total Volume'], q=4,
                                 labels=['Poor', 'Modest', 'Good', 'Excellent'])
```

```python
avocado.head()
```

```python
sns.countplot(data=avocado, x='year', hue='Sales Volume')
```

```python
bag_sizes = avocado.melt(
    id_vars='year',
    value_vars=['Small Bags','Large Bags', 'XLarge Bags'],
    var_name='Bags', value_name='Count')
```

```python
bag_sizes
```

```python
sns.relplot(data=bag_sizes, kind='line', x='year', y='Count', hue='Bags')
```
